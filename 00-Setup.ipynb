## Install Packages
 > Usually need to run this only once or when a new package needs to be installed.

```python
# Install required packages
import subprocess
import sys

packages = [
    'dotenv',
    'langchain',
    'langchain-community',
    'langchain_openai',
    'langchain_anthropic',
]

print("Installing required packages...")
for package in packages:
    subprocess.check_call([sys.executable, '-m', 'pip', 'install', package])

print("✅  All packages installed successfully.")
print("=" * 70)
```
